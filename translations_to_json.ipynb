```python
import pandas
import json
```

```python
df = pandas.read_csv('./data/korean_subtitles_translation.csv')
```

```python
df.head()
```

```python
col = df.columns[1:]
```

```python
group_df = []
for vid, group in df.groupby('video'):
    srt = group[col].to_dict(orient='records')
    group_df.append({
        'video': vid,
        'srt': srt
    })
```

```python
with open('./data/korean_subtitles.json', 'w') as json_file:
    json.dump(group_df, json_file, indent=4)
```
